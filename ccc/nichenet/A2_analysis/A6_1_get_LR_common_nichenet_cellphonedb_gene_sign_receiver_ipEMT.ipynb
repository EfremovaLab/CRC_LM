## Only select in cellphonedb ligand-receptor interactions that are found in nichenet analysis
this is done to get receptors for the ligands 

1. get interactions that are in cellphonedb as well
2. find ligands that are missing and get receptor interactions from the nichenet analysis that are not found in cellphonedb
3. get potential of interaction ligand - receptor from nichenet to be able to chose 1 receptor per ligand
4. create single df with both cellphonedb and nichenet interactions

----
### 1. get interactions that are in cellphonedb as well

```python
# create interacting pair L-R for looking into same interaction in cellphonedb
nichenet_lr["interacting_pair"] = nichenet_lr["from"]+"_"+nichenet_lr["to"]
nichenet_lr["interacting_pair"]
```

```python
# create interacting pair R-L for looking into same interaction in cellphonedb
nichenet_lr["interacting_pair_rev"] = nichenet_lr["to"]+"_"+nichenet_lr["from"]
```

---

```python
from_to = [f"{s}|{receiver}" for s in senders]
```

### B. get interactions responder (to) | sender (from) with `nichenet_lr["interacting_pair_rev"]`

```python
to_from = [f"{receiver}|{s}" for s in senders]
to_from
```

```python
columns2select_rev = ["id_cp_interaction", "interacting_pair", "partner_a", "partner_b", "gene_a", "gene_b",
                  "secreted", "receptor_a" , "receptor_b"] + to_from 
```

```python
to_from_df = sign_means.loc[sign_means.interacting_pair.isin(nichenet_lr["interacting_pair_rev"]), columns2select_rev]
to_from_df.shape
```

```python
to_from_df
```

```python
# drop interactions that are nan here 
to_from_df.dropna(axis=0, how="all", subset=to_from, inplace=True)
to_from_df.shape
```

### C. concatenate these interactions into single dataframe 
* but first need to reverse the interaction from to_from so we have same direction as from_to (Ligand-Receptor)
* create `from` and `to` columns in both dfs and then we extract and concatenate these 

```python
to_from_df["from"] = [ele[1] for ele in to_from_df.interacting_pair.str.split("_")]
to_from_df["to"] = [ele[0] for ele in to_from_df.interacting_pair.str.split("_")]
```

```python
from_to_df["from"] = [ele[0] for ele in from_to_df.interacting_pair.str.split("_")]
from_to_df["to"] = [ele[1] for ele in from_to_df.interacting_pair.str.split("_")]
```

```python
cpdb_lr_df = pd.concat([from_to_df[["from", "to", "interacting_pair", "id_cp_interaction"]], 
                        to_from_df[["from", "to", "interacting_pair", "id_cp_interaction"]]
                       ])
cpdb_lr_df.shape
```

### B. get receptor for each ligand in nichenet (that hasn't been matched by cellphonedb)
* list of interactions from nichenet (ligands are present in cellphonedb)

```python
len(vis_lr_nichenet.columns)
```

```python
toadd = list()
for ligand in ligands2add:
    # select interactions and order by potential (decreasing)
    tmp = vis_lr_nichenet.loc[vis_lr_nichenet.loc[:,ligand]>0, ligand].sort_values(ascending=False)
    tmp = tmp.reset_index().copy()
    tmp.rename(columns={"index": "to", ligand:"potential"}, inplace=True)
    tmp["from"] = ligand
    toadd.append(tmp)

# concat all so we get a df with interactions, NOPPE messes everything up if we sort by ligand!
nichenet_lr_missing = pd.concat(toadd)
```

```python
nichenet_lr_missing
```

#### One way would be to keep the receptor with the highest interaction potential, but we could also keep more than 1 receptor
if keeping 1 recepotor, use `drop_duplicates(subset=["from"], keep="first")`

#### Otherwise double sort the columns by ligand and then by potential!!! to keep all ligands but in sorted order

```python
# keep only first interaction for each ligand (as they are already ordered by interaction potential)
#nichenet_lr_missing_filtered = nichenet_lr_missing.drop_duplicates(subset=["from"], keep="first").copy()
```

```python
# list of interactions from nichenet (ligands are present in cellphonedb)
nichenet_lr_missing_filtered = nichenet_lr_missing.sort_values(by=["from", "potential"], ascending=[True, False]).copy()
nichenet_lr_missing_filtered
```

```python
nichenet_lr_missing_filtered.shape
```

### C. check that these interactions are not in cellphonedb
* check means file and find pair of ligands-recerptor that is in nichenet and cellpphonedb but is not significant in cellpphonedb

```python
# check that these interactions are not in cellphonedb
# load significant means result from cellphonedb
means = pd.read_csv("/data/BCI-CRC/nasrine/data/CRC/spatial/CRC_LM_VISIUM/CRC_LM_VISIUM_04_08_09_11/cellphonedb3/concat_withWu2022/cpdb3_output/means.txt",
                         sep="\t"
                        )

check1 = nichenet_lr_missing_filtered["from"] + "_" + nichenet_lr_missing_filtered["to"]
check2 = nichenet_lr_missing_filtered["to"] + "_" + nichenet_lr_missing_filtered["from"]

ligands2discard_df = means.loc[(means.interacting_pair.isin(check1)) | (means.interacting_pair.isin(check2)), :]

ligands2discard_df
```

```python
ligands2discard = list(ligands2discard_df["gene_a"].drop_duplicates()) + list(ligands2discard_df["gene_b"].drop_duplicates())
print("We have to remove {} genes (some could be receptors so they won't even be in it) from ligands common as they are not significant in cellphonedb and exist in cpdb".format(len(ligands2discard)))
```

```python
ligands2discard
```

```python
nichenet_lr_missing_filtered["interaction"] = nichenet_lr_missing_filtered["from"] + "_" + nichenet_lr_missing_filtered["to"]
nichenet_lr_missing_filtered["interaction_rev"] = nichenet_lr_missing_filtered["to"] + "_" + nichenet_lr_missing_filtered["from"]
```

```python
# remove interaction that is not significant in cpdb but exists in cpdb means.txt
nichenet_lr_missing_filtered = nichenet_lr_missing_filtered[~
    ((nichenet_lr_missing_filtered.interaction.isin(ligands2discard_df["interacting_pair"])) | (
        nichenet_lr_missing_filtered.interaction_rev.isin(ligands2discard_df["interacting_pair"])))].copy()

nichenet_lr_missing_filtered.shape
```

---
### 3. for cellphonedb interactions, also get nichenet potential to keep 1 receptor 

```python
get_potential_df = list()
for ligand in set(cpdb_lr_df["from"]):
    # select interactions and order by potential (decreasing)
    tmp = vis_lr_nichenet.loc[vis_lr_nichenet.loc[:,ligand]>0, ligand].sort_values(ascending=False)
    tmp = tmp.reset_index().copy()
    tmp.rename(columns={"index": "to", ligand:"potential"}, inplace=True)
    tmp["from"] = ligand
    get_potential_df.append(tmp)
```

```python
# concat all so we get a df with interactions, NOPPE messes everything up if we sort by ligand!
get_potential_df = pd.concat(get_potential_df)
get_potential_df.head(5)
```

```python
# append potential calculated by nichenet to interactions that are present in cellphonedb to filter out receptors
# and keep only 1 receptor per ligand
cpdb_lr_df_potential = cpdb_lr_df.merge(right=get_potential_df, 
                                        how="left",
                                        left_on=["to", "from"],
                                        right_on=["to", "from"],
                                        validate="one_to_one")

# drop duplicates so we only keep 1 interaction per ligand that has the receptor with highest potential 
#cpdb_lr_df_potential.drop_duplicates(subset=["from"], keep="first", inplace=True)

# sort by potential so get receptors with highest potenial first 
cpdb_lr_df_potential.sort_values(by=["from", "potential"], ascending=[True, False], inplace=True)

```

```python
cpdb_lr_df_potential
```

```python
cpdb_lr_df_potential.shape
```

---
### 4. create single df of interactions

```python
final_lr = pd.concat([cpdb_lr_df_potential[["from", "to", "potential"]],
                      nichenet_lr_missing_filtered[["from", "to", "potential"]]
                     ], axis=0)
final_lr.shape
```

```python
final_lr.sort_values(by=["from", "potential"], ascending=[True, False], inplace=True)
```
